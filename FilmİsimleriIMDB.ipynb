```python
import requests
from bs4 import BeautifulSoup
url='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response=requests.get(url)
print(response)
```

```python
content=response.content#incele kısmına ulaşma
print(content)
```

```python
soup=BeautifulSoup(content,"html.parser")
list=soup.find('tbody',{'class':'lister-list'}).find_all('tr',limit=50)#tbody altındaki trleri alacağız. class="lister-list" 
list
sayac=0
```

```python
for tr in list:
    baslik=tr.find('td',{'class':'titleColumn'}).find('a').text##ilk filmin ismi
    #print(baslik)
    yil=tr.find('td',{'class':'titleColumn'}).find('span').text.strip('()')##filmin yılı
    #print(yil)
    reyting=tr.find('td',{'class':'ratingColumn imdbRating'}).find('strong').text##filmin rating oranı
    #print(reyting)
    print(f'{sayac}-Film Adı:{baslik.ljust(50)} , Yılı:{yil} , Değerlendirmesi:{reyting}')##ljust yer ayırma string için sola yasla
    sayac+=1
```
